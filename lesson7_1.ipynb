```python
# 90(含)~100分為「優」
# 80(含)~89分為「甲」
# 70(含)~79分為「乙」
# 60(含)~69分為「丙」
# 0(含)~59分為「丁」
import pyinputplus as pyip
scores = pyip.inputInt(prompt="請輸入分數:",min=0,max=100)
print(scores)
message = ""
#巢狀判斷,4個if
if scores >= 90:
    message = "優"
else:
    if scores >= 80:
        message = "甲"
    else:
        if scores >= 70:
            message = "乙"
        else:
            if scores >= 60:
                message = "丙"
            else:
                message = "丁"
print(message)
```

```python
# 輸入顧客購買金額，若金額在
# 100000元打8折. 
# 50000打85折. 
# 30000打9折. 
# 10000打95折. 

# 請輸入購買金額:130000
# 實付金額是: 104000.0 元
import pyinputplus as pyip
money = pyip.inputInt(prompt="請輸入顧客購買金額:",min=0)
print("購買金額:", money)

#多項選擇,1個if
if money >= 1000000:
    money *= 0.8
elif money >= 50000:    
    money *= 0.85
elif money >= 30000:     
    money *= 0.9
elif money >= 10000:     
    money *= 0.95
else:
    money *= 1
print("實付金額是:", int(money), "元")
```

```python
# 輸入顧客購買金額，若金額在
# 滿千送百再打折
# 100000元打8折. 
# 50000打85折. 
# 30000打9折. 
# 10000打95折. 

# 請輸入購買金額:130000
# 實付金額是: 104000.0 元
import pyinputplus as pyip
money = pyip.inputInt(prompt="請輸入顧客購買金額:",min=0)
print("購買金額:", money)
Discount = money//1000*100
print ("滿千折扣金額", Discount)
Pmoney= money-Discount

#多項選擇,1個if
if Pmoney >= 100000:
    Pmoney *= 0.8
elif Pmoney >= 50000:    
    Pmoney *= 0.85
elif Pmoney >= 30000:     
    Pmoney *= 0.9
elif Pmoney >= 10000:     
    Pmoney *= 0.95
else:
    Pmoney *= 1
print("實付金額是:", int(Pmoney), "元")
```
